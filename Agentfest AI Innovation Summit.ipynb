### Metric Views
The **SEMANTIC LAYER** that sits right next to your data and acts as the **SOURCE OF TRUTH** for all downstream data assets

![Metric View](/Volumes/hr_molina_workshop/appeals_review/img/metric_view.png)

#### Prompt: <br>
Considering the above business requirements, go through the tables in **hr_molina_workshop.appeals_review** schema and create a metric view. I want all of these:

##### Measures

| # | Name | Expression | Source Table | Description |
|---|---|---|---|---|
| 1 | Active Member Count | `COUNT_IF(is_active = true)` | `eligibility` | Members with coverage right now |
| 2 | New Enrollments | `COUNT(eligibility_id)` | `eligibility` | Coverage starts in a given period |
| 3 | Active WA Medicaid Members | `COUNT(DISTINCT state_member_id) WHERE eligibility_status = 'Active'` | `wa_hca_eligibility` | WA Medicaid headcount for PMPM base |


##### Dimensions

| # | Name | Expression | Source Table | Description |
|---|---|---|---|---|
| 1 | Plan Type | `plan_type` | `eligibility` | Commercial, Medicaid, Medicare, CHIP |
| 2 | WA County | `county` | `wa_hca_eligibility` | King, Pierce, Clark, Spokane, etc. |
| 3 | Enrollment Month | `DATE_TRUNC('MONTH', coverage_start)` | `eligibility` | Month-over-month trend slicing |

Also, identify any other useful metrics and dimensions that I may have missed.
